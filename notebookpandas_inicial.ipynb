# Manipulando os dados

## Criando colunas numéricas

```python
#valor_por_mes: essa coluna deve conter os gastos mensais de cada imóvel, incluindo aluguel e condomínio;

#astos anuais por imóvel, ou seja, IPTU mais 12 meses de aluguel e condomínio

url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep=';')
dados.head()
```

```python
#começar a criar a coluna valor por mes
# coloca o nome do data frame que é dados e coloca uma coluna ja existente
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
dados.head()
#valor do aluguel ue é valor mais condominio, isso para encontrar os gastos do mes entao soma esse valors
```

```python
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
dados.head()
#então aparece na ultima coluna uma nova coluna chamada valor por mes
```

```python
#agora cria a coluna valor por ano
dados['Valor_por_ano'] = dados['Valor_por_mes'] * 12 + dados['IPTU']
dados.head()
```

## Criando colunas categóricas

```python
#Descrição: essa coluna deve possuir uma sumarização das principais informações dos imóveis que serão apresentadas no site: tipo de imóvel, bairro, quantidade de quartos e vagas de garagem;

#Possui_suite: essa deve ser uma coluna que informe apenas se o imóvel possui ou não suítes, sem se importar com a quantidade.

#colocar em duas colunas ja concatenadas
dados['Descricao'] = dados['Tipo'] + ' em ' +dados ['Bairro']
#coluna tipo e bairro sao strings então se concatena e nao soma. se somar a visualização fica estranha
#dados['Descricao'] = dados['Tipo'] + dados ['Bairro']
dados.head()
```

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados ['Bairro'] + ' com' + \
dados['Quartos'] + ' quarto(s) ' + \ ' e ' + dados['Vagas'] + ' vaga(s) de garagem'
dados.head()
#barra invertida quebrando a linha
#nao se concatena stringe com inteiro, converter quartos e coluna de garagem com o metodo stype
```

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados ['Bairro'] + ' com ' + \
                     dados['Quartos'].astype(str) + ' quarto(s) ' + \
                     ' e ' + dados['Vagas'].astype(str)  + ' vaga(s) de garagem'
dados.head()
```

```python
#colina possui suite analisando a coluna suites
#analisa cada illhnha, caso tenha maior que zero fica sim, do contrario nao
#apply aplica determinada funcao em conjunto de dados - aqui é suites, usando labda do puthon que sao personalizadas escritas em uma unica linha de codigo
#se o x que é o registro for maior que zero coloca sim
dados['Possui_susite'] = dados['Suites'].apply(lambda x: 'Sim' if x > 0 else 'Não')
dados.head()
```

```python
dados.to_csv('dados_completos_dev.csv',index=False, sep=';')
```
